```python
import pandas as pd
```

```python
dane1 = pd.read_csv( '../dane/2022_I_kw/Wyniki/93030__20220101T000000__20220401T000000.csv' )
```

```python
dane2 = pd.read_excel( '../dane/kzw/2022_1_kw.xlsx' )
```

```python
dane2.head()
```

```python
dane = dane2[ dane2['ID_GPS'] == 93030 ]
```

```python
dane_1 = dane.iloc[[1]]
dane_1
```

```python
dane
```

```python
dane_1
```

```python
# Zamiana czasu na format datetime
dane_1['Czas wyjazdu ZRM'] = pd.to_datetime(dane_1['Czas wyjazdu ZRM'])
dane_1['Czas powrotu ZRM'] = pd.to_datetime(dane_1['Czas powrotu ZRM'])
dane1['GPS_TIME'] = pd.to_datetime(dane1['GPS_TIME'])

# Połączenie tabel na podstawie warunku czasowego
merged_data = pd.merge_asof(
    dane1.sort_values('GPS_TIME'), 
    dane_1.sort_values('Czas wyjazdu ZRM'), 
    left_on='GPS_TIME', 
    right_on='Czas wyjazdu ZRM', 
    direction='backward'
)

# Filtracja tylko rekordów, gdzie GPS_TIME mieści się w przedziale wyjazdu-powrotu
merged_data = merged_data[
    (merged_data['GPS_TIME'] >= merged_data['Czas wyjazdu ZRM']) & 
    (merged_data['GPS_TIME'] <= merged_data['Czas powrotu ZRM'])
]

merged_data
```
